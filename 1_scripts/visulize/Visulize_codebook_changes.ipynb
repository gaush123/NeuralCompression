```python
%matplotlib inline
import sys, os
import matplotlib.pyplot as plt 
import pickle
import numpy as np
caffe_root = os.environ["CAFFE_ROOT"]
os.chdir(caffe_root)
dir_t = '2_results/kmeans/alexnet/'
```

```python
num = 50
codebook=[0]*num
for i in range(num):
    snapshot = dir_t + 'codebook%d'%i 
    codebook[i] = pickle.load(open(snapshot))
layers = codebook[0].keys()
```

```python

magnitude_diff = np.zeros((len(layers)))
for i in range(num-1):
    for idx, layer in enumerate(layers):
        magnitude_diff[idx] += np.sqrt(np.mean((codebook[i+1][layer]-codebook[i][layer]) ** 2)) / np.sqrt(np.mean(codebook[i][layer] ** 2))
```

```python
plt.plot(range(len(layers)), magnitude_diff)
plt.show()
```
